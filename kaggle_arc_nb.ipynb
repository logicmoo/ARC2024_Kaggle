```python

from pyswip import Prolog
prolog = Prolog()
prolog.retractall("knows(michael,_)")
prolog.assertz("knows(michael,john)")

for soln in prolog.query("knows(X,Y)"):
    print(soln["X"], "knows of", soln["Y"])

from IPython.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))


for soln in prolog.query("ignore(cd(kaggle_arc))"): print()

for soln in prolog.query("consult(kaggle_arc)"): print()
```

```python
# Imports for JupyterLite
# %pip install -q ipywidgets
import ipywidgets as widgets
#print(widgets.Button.on_click.__doc__)
from IPython.display import display, clear_output


out = widgets.Output(layout={'border': '1px solid black'})
out.clear_output()

boutput = out # widgets.Output()

output = widgets.Output()
output.clear_output()

def make_action_button(str):
    button = widgets.Button(description=str)
    display(button, boutput)
    def on_button_clicked(b):
        for soln in prolog.query(f"{str},!"): print()        
    button.on_click(on_button_clicked)

make_action_button("print_test")

make_action_button("ls")

display(output, boutput)
output.clear_output()
```
